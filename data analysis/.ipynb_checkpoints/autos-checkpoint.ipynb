```python
url = "imports-85.data"
df = pd.read_csv(url, header=None)
headers = ["symboling","normalizes-loses","make","fuel-type","aspiration","no-of-doors","body-style",
           "drive-wheels","engine-location","wheel-base","length","width","height","curb-weight","engine-type",
           "no-of-cylinders","engine-size","fuel-system","bore","stroke", "compression-ratio","horsepower",
           "peak-rpm","city-mpg","highway-mpg","price"]
df.columns = headers
df.head()
```

```python
# # HOw to drop misiing values

#         dataframes.dropna()
# axis =0 drops the entire row
# axis =1 drops the entire column


# df.dropna(subset=[('normalizes-loses')], axis=0, inplace=True)
df.isna()
# df.notna()

df['normalizes-loses'] = pd.to_numeric(df['normalizes-loses'], errors='coerce')
df['price'] = pd.to_numeric(df['price'], errors='coerce')

# df = df[~((df['normalizes-loses'] == '?' ))].astype(float)

df
```

```python
# how to replace missing values?

# df.replace(misiing_value, new_value)

mean = df["normalizes-loses"].mean()
print("The mean of normalizes-loses is:", mean)
df["normalizes-loses"]= df["normalizes-loses"].replace(np.nan, mean)
df
```

```python
# # data formatting
# #     bringing data into a commom standard of expression allows users to make mesningful comparisons
#         For consistency and understanable


#converting "mpg" to L/100Km

df["city-mpg"] = 235/df["city-mpg"]

df.rename(columns={"city-mpg": "city-L/100km"}, inplace=True)
```

```python
df.dtypes 
```

```python
mean = df["price"].mean()
print("The mean of prices is:", mean)
df["price"].replace(np.nan, mean)


df["price"] = df["price"].astype("int64")
df
```

```python
df["length"]
```

```python
# # Data Normalization

# Uniform the featues values with different range

# 1. simple feature scaling 
#             x-new = x-old / x-max

# 2. min-MAx
#             x-new = (x-old - x-min) / (x-max -x-min)

# 3. Z-score
#             x-new = (x-old - mean) / S.D
#             ranges from -3 to 3 id if data normally distributed
```

```python
# Applying MIN_MAX 
df["length"] = (df["length"] - df["length"].min())/(df["length"].max()-df["length"].min())
df["length"]
```

```python
# data bining

bins = np.linspace(min(df["price"]),max(df["price"]),4)
group_names = ["Low", "Medium", "High"]
df["price-binned"] = pd.cut(df["price"], bins, labels=group_names, include_lowest=True)
df
```
